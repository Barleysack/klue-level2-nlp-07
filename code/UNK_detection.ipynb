## 목표 : UNK을 찾아내는 함수를 구현하고 Synonym 대량 작업을 해보자

```python
MODEL_NAME = 'klue/roberta-large'
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
data = load_data("../dataset/train/train.csv")
data.sample(10)
```

```python
data[data.subject_entity=="'신라대학교'"]
```

```python
data[data.subject_entity=="'스페인'"]
```

```python
# 이걸 함수화해서 load_data.py에 반영해야됨
# 숫자 사이에 있는 콤마(,)에서 콤마 지우는 방법
data['sentence'] = data['sentence'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data['subject_entity'] = data['subject_entity'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data['object_entity'] = data['object_entity'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data[data.subject_entity=="'스페인'"]
```

```python
print(tokenizer.decode(tokenizer.encode('6000년')))
print(tokenizer.encode('6000년'))
print(tokenizer.decode(tokenizer.encode('6,000년')))
print(tokenizer.encode('6,000년'))
```
