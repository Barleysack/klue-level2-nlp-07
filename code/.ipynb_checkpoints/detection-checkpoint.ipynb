## 숫자 사이 콤마를 어떻게 처리할 것인가

```python
MODEL_NAME = 'klue/roberta-large'
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
data = load_data("../dataset/train/train.csv")
data.sample(10)
```

```python
# 이걸 함수화해서 load_data.py에 반영해야됨
# 숫자 사이에 있는 콤마(,)에서 콤마 지우는 방법
data['sentence'] = data['sentence'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data['subject_entity'] = data['subject_entity'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data['object_entity'] = data['object_entity'].apply(lambda x: re.sub(r'(\d+),(\d+)', r'\1\2', x))
data[data.subject_entity=="'스페인'"]
```

```python
eda = EDA(morpheme_analyzer="Mecab", alpha_sr=-0.8, alpha_ri=0, alpha_rs=0, prob_rd=0)
aeda = AEDA(morpheme_analyzer="Mecab", punc_ratio=0.15, punctuations=["?", "!", ".", ",", "또한", "#"])

def right_augment(en1, en2, text):
    if en1 not in text or en2 not in text:
        return True
    return False

def augment_sent(df, my_eda, my_aeda):
    sentences = list(df['sentence'])
    entities1, entities2 = list(df['object_entity']), list(df['subject_entity'])
    aug_sentences = []
    for i, sent in enumerate(sentences):
        aug_sent = aeda(sent)
        ent1, ent2 = entities1[i], entities2[i]
        if not right_augment(ent1, ent2, aug_sent):
            aug_sent = sent
        aug_sentences.append(aug_sent)
    df['sentence'] = aug_sentences
    return df
```

```python
sample_df = data.sample(100)
aug_df = augment_sent(sample_df, eda, aeda)
```
