```python
lst1 = list(train1['sentence'])
lst2 = list(train2['sentence'])
lst3 = list(train3['sentence'])
differ = []
for i in range(len(train1)):
    if lst1[i] != lst2[i] or lst2[i] != lst3[i] or lst3[i] != lst1[i]:
        differ.append(i)
len(differ)
```
